```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
data = pd.read_csv('/content/Final_TrainData_ACL&ALL.csv') #Training/labeled data file 
data.head()
```

# Visulising the Annotation result from JASST, IPM ,and ACL

```python
data['Labels'] = data['Labels'].apply(lambda x: " ".join(x.upper() for x in x.split()))
data['Labels'].describe()
```

```python
print(data['Labels'].value_counts())
```

```python
for Sour, Count in enumerate(data['Source'].value_counts().index.tolist()):
    print('Contributions',  Count)
    print('Counts :', (data['Labels'].value_counts()[Sour]))
```

```python
data['Labels'] = data['Labels'].apply(lambda x: " ".join(x.upper() for x in x.split()))
data['Labels'].value_counts().plot(kind='barh', figsize=(10, 4), rot=0)
plt.title("The Distribution of Research Contributions From IPM, JASIST,and ACL In All Categories ", y=1.02)
```

```python
Colors=['red', 'yellow','blue', 'green',  'pink', 'gray','violet', 'orange','maroon']

Data3 = data['Labels'].value_counts().keys().tolist()
Valcount = data['Labels'].value_counts().tolist()
plt.pie(Valcount,labels=Data3, colors=Colors, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

# IPM Visualization 

```python
Colors=['red', 'yellow','blue', 'green',  'pink', 'black','violet', 'orange','azure']
IPM_data= pd.read_csv('/content/IPM_Labels.csv')
IPM_data['Labels'] = IPM_data['Labels'].apply(lambda x: " ".join(x.upper() for x in x.split()))
IPM_data['Labels'].count()
```

```python
import matplotlib.pyplot as plt
% matplotlib inline
import seaborn as sns
plt.figure (figsize = (10,4)) 
IPM = sns.countplot (y = IPM_data['Labels'],data=IPM_data, order=IPM_data['Labels'].value_counts().index)
plt.title("The Distribution of Research Contributions From IPM", y=1.02)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

IPM_Value = IPM_data['Labels'].value_counts().keys().tolist()
IPM_Count = IPM_data['Labels'].value_counts().tolist()
plt.pie(IPM_Count, labels=IPM_Value, autopct='%1.1f%%')
plt.axis('equal')
plt.show()

```
